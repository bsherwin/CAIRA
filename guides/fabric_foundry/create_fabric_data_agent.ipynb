Uses NYC taxi data.

```python
from fabric.dataagent.client import (
    FabricDataAgentManagement,
    create_data_agent,
    delete_data_agent,
)

# Hardcoded for now
data_agent_name = "data_agent_automation_sample"
lakehouse_name = "NYCTaxi_756"

# create DataAgent
data_agent = create_data_agent(data_agent_name)

data_agent.update_configuration(
    instructions="You are a helpful assistant, help users with their questions",
)

# datasource type could be: lakehouse, kqldatabase, warehouse or semanticmodel
data_agent.add_datasource(lakehouse_name, type="lakehouse")

# Publish
data_agent.publish()

datasource = data_agent.get_datasources()[0]

# Add data sources
datasource.select("dbo", "year_2017")
datasource.select("dbo", "year_2018")
datasource.select("dbo", "year_2019")
datasource.select("dbo", "year_2020")

# Add additional instructions to the datasource
ds_notes = "When asked about taxi ridership information use tables in dbo with year_ prefix to retrieve detailed information"
datasource.update_configuration(instructions=ds_notes)

# Get configuration
datasource.get_configuration()
```
